```python
# --- Lab 3.1 Part 2: Full 12x12 Multiplication Grid ---
print("12x12 Multiplication Table:")
for i in range(1,13): #เติมตัวแปรให้สมบูรณ์
    for j in range(1, 13):
        print(f"{i*j:4}", end="")   # จัดช่องกว้าง 4 ตัวอักษร
    print()   # ขึ้นบรรทัดใหม่เมื่อจบแต่ละแถว

print()  # spacer
```

```python
# --- Lab 3.2 Part 2: Guessing Game ---

import random
numrand = int((random.random())*100)

heart = 5
n_guess = 1
print("====Guessing Game====")

while heart != 0:
    print(f"=== Guess {n_guess} ===")
    guess = int(input("Guess the number: "))
    try:
        guess = int(guess)
        if guess == numrand:
            print("Congratulations! You guessed it!")
            print(f"The number is {numrand}")
            break
        elif guess > numrand:
            print(f"{guess} is too high! Try again.")
            heart -= 1
            print(f"You have {heart} heart(s) left")
        elif guess < numrand:
            print(f"{guess} is too low! Try again.")
            heart -= 1
            print(f"You have {heart} heart(s) left")
        
        if heart == 0:
            print("You are LOSE!!!!!")
            print(f"The number is {numrand}")
        n_guess += 1
    except ValueError:
        print("Your guess must be the number")

    print()
```

### ***Activity 3.1*** - *Debugging Infinite Loops & Loop Logic Errors*

***Format***: Individual or pair work, followed by class discussion/demonstration.

***Instructions***: Provide students with several buggy code snippets containing:
-  An infinite while loop (missing update of loop control variable).
-  A for loop that doesn't produce the desired range (incorrect range() arguments).
-  A loop where break or continue is used incorrectly.
-  Students must identify the bug, explain why it happens, and fix the code.

***Purpose***: Develop critical debugging skills, understand common loop pitfalls.

```python
# Buggy Snippet A: Infinite Loop
# counter = 0
# while counter < 5:
#     print(counter)
#     # Missing counter increment!

# Answer
counter = 0
while counter < 5:
    print(counter)
    counter += 1
```

```python
# Buggy Snippet B: Incorrect Range
# for i in range(1, 5): # Goal: print 1, 2, 3, 4, 5
#     print(i)

# # Answer
for i in range(1, 6): # Goal: print 1, 2, 3, 4, 5
    if i < 5:
        print(f"{i}", end=", ")
    else:
        print(f"{i}", end="")
```

```python
# Buggy Snippet C: Misplaced Break
# for char in "hello":
#     if char == 'l':
#         print("Found 'l', stopping!")
#         break
#     print(char) # This will print 'h', 'e', 'l' then break

# Answer
for char in "hello":
    print(char)
    if char == 'l':
        print("Found 'l', stopping!")
        break
```

#### ***Activity 3.2*** - *Pattern Printing with Nested Loops*

***Format***: Individual creative coding challenge.

***Instructions***: Ask students to write Python code using nested for loops to print simple ASCII art patterns.

```python
# A square of asterisks:
# *****
# *****
# *****
# *****
# *****

inp = input("Enter Square Range (e.g. 5): ")
try:
    inp = int(inp)
    print(f"A square of asterisks {inp} radius:")
    for i in range(1,inp):
        print("*"*inp)
except:
    print("Must be Number")
```

```python
# A right-angled triangle of asterisks:
# *
# **
# ***
# ****
# *****

inp = input("Enter Square Range (e.g. 5): ")
try:
    inp = int(inp)
    print(f"A simple pyramid of asterisks {inp} radius:")
    for i in range(1,inp+1):
        print("*"*i)
except:
    print("Must be Number")
```

### ***Challenge*** - An inverted right-angled triangle or a pyramid.
**Purpose**: Reinforce understanding of nested loops and the relationship between loop variables and printed output. Encourages algorithmic thinking and problem-solving.


```python
# A right-angle pyramid

inp = input("Enter Square Range (e.g. 5): ")
try:
    print(f"A right-angle pyramid of asterisks {inp} radius:")
    inp = int(inp) + 1 
    for i in range(1,inp):
        print(" "*(inp-i)+"*"*(i))
except:
    print("Must be Number")
```

```python
# A middle-angle pyramid

inp = input("Enter Square Range (e.g. 5): ")

try:
     inp = int(inp)
     print(f"A middle-angle pyramid of asterisks {inp} radius:")

     for i in range(1, inp + 1):
          print(" "*10,(" " * (inp - i) + "".join(["*"] * (2 * i - 1))), " "*10)

except:
     print("Must be Number")
```
