```python
# --- Lab 3.1 Part 1: Single Number Multiplication Table ---
n = int(input("Enter a number for its multiplication table (e.g., 7): "))
for i in range(1, 13):
    print(f" {n}x{i} = {i*n}")
print()  # spacer
```

```python
# --- Lab 3.1 Part 2: Full 12x12 Multiplication Grid ---
print("12x12 Multiplication Table:")
for i in range(1,13): #เติมตัวแปรให้สมบูรณ์
    for j in range(1, 13):
        print(f"{i*j:4}", end="")   # จัดช่องกว้าง 4 ตัวอักษร
    print()   # ขึ้นบรรทัดใหม่เมื่อจบแต่ละแถว

print()  # spacer
```

```python
# --- Lab 3.2 Part 1: Countdown Timer ---
import time

start = int(input("Enter starting number for countdown: "))
while start >= 0:
    time.sleep(1)
    print(start)
    start -=1
    #คำใบ้ print ออกมาในรูปแบบ Countdown: {start} และลดค่าของ start ลง 1
```

```python
# --- Lab 3.2 Part 2: Guessing Game ---

import random
numrand = int((random.random())*100)

heart = 5
n_guess = 1
print("====Guessing Game====")

while heart != 0:
    print(f"=== Guess {n_guess} ===")
    guess = int(input("Guess the number: "))
    try:
        guess = int(guess)
        if guess == numrand:
            print("Congratulations! You guessed it!")
            print(f"The number is {numrand}")
            break
        elif guess > numrand:
            print(f"{guess} is too high! Try again.")
            heart -= 1
            print(f"You have {heart} heart(s) left")
        elif guess < numrand:
            print(f"{guess} is too low! Try again.")
            heart -= 1
            print(f"You have {heart} heart(s) left")
        
        if heart == 0:
            print("You are LOSE!!!!!")
            print(f"The number is {numrand}")
        n_guess += 1
    except ValueError:
        print("Your guess must be the number")

    print()
```
