## **Activity 2.1** : Trace Execution of Conditional Code Snippets

***Format***: Individual work with peer discussion/instructor led review.

***Instructions***: Provide several code snippets with if/elif/else and logical operators. Students must trace the execution step-by-step for given input values and determine the final output.

***Purpose***: Reinforce understanding of how conditions are evaluated and how the program flow changes.


```python
# Snippet A
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"
print(f"Grade: {grade}")



# What's the output?

## จะ print "Grade: B"
## EZ ค้าบบบบบ
```

```python
# Snippet B
temp = 25
is_raining = True
if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")



# What's the output?

## จะ print "Stay indoors or bring an umbrella."
## ตรรกศาสตร์มันง่ายคับบบ
```

```python
# Snippet C
age = 17
has_license = True
if age >= 18:
    if has_license:
        print("You can drive legally.")
    else:
        print("You need a license to drive.")
else:
    print("You are too young to drive.")



# What's the output?

## จะ print "You are too young to drive."
## ขอยากๆหน่อยดิ้
```

## **Activity 2.2**: Design a Simple "Choose Your Own Adventure" Text Game

***Format***: Small groups (2-3 students).

***Instructions***:
Brainstorm: As a group, come up with a very short "choose your own adventure" story concept (e.g., entering a haunted house, exploring a forest).

*Design*: Outline 2-3 decision points in your story. For each decision, determine 2-3 possible choices and their outcomes.

*Pseudo-code/Outline*: Write down the basic input() prompts and the if-elif-else structure needed to implement your story's decisions. You don't need to write full Python code yet, just the logic.

*Share*: Each group presents their story concept and their conditional logic outline to the class.


***Purpose***: Apply conditional logic to a creative, problem-solving scenario, fostering design thinking and collaboration before coding.


```python
# Idea
"""
เกมทายตัวเลขโดยตัวเกมจะสุ่มเลขให้เราจาก 1 - 100
แล้วให้เราทายไปเรื่อยๆจนกว่าจะถูกต้อง
ซึ่งผู้เล่นจะทายได้แค่ 5 ครั้ง

โดยสมมุต
เกมสุ่มได้เลข 53
เราทาย 56 >> เกมจะบอกเราว่า 56 เป็นเลขที่สูงไป
เราทายอีกครั้ง 47 >> เกมจะบอกว่า 57 เป็นเลขที่ต่ำไป

หากตัวเลขูกต้อง ก็ชนะ หากยังผิด 5 ครั้งก็แพ้
"""
```

```python
# Raw Coding
import random
numrand = int((random.random())*100)

heart = 5
guess = 1
print("====Guess Number Game====")

while heart != 0:
    print(f"=== Guess {guess} ===")
    user_input = input("Enter your guess: ")
    try:
        user_input = int(user_input)
        if user_input == numrand:
            print("You are WIN!!!!!")
            print(f"The number is {numrand}")
            break
        elif user_input > numrand:
            print(f"{user_input} is higher than the number")
            heart -= 1
            print(f"You have {heart} heart(s) left")
        elif user_input < numrand:
            print(f"{user_input} is lower than the number")
            heart -= 1
            print(f"You have {heart} heart(s) left")
        elif heart == 0:
            print("You are LOSE!!!!!")
            print(f"The number is {numrand}")
        guess += 1
    except ValueError:
        print("Your guess must be the number")

    print()
```

```python
# Coding more beautiiful by ChatGPT
import random
numrand = int((random.random())*100)

heart = 5
guess = 1

print("=" * 35)
print("🎯      GUESS NUMBER GAME      🎯")
print("=" * 35)
print("Guess the number between 1 - 100")
print(f"❤️ Hearts: {heart}")
print()

while heart > 0:
    print("-" * 35)
    print(f"🎲 Guess #{guess}")
    print("-" * 35)

    user_input = input("👉 Enter your guess: ")

    try:
        user_input = int(user_input)

        if user_input == numrand:
            print("\n🎉 CONGRATULATIONS! YOU WIN! 🎉")
            print(f"🏆 The number was: {numrand}")
            break

        elif user_input > numrand:
            heart -= 1
            print(f"📈 {user_input} is TOO HIGH!")

        else:
            heart -= 1
            print(f"📉 {user_input} is TOO LOW!")

        if heart > 0:
            print(f"❤️ Hearts left : {heart}")
            print("🔄 Try again!\n")
        else:
            print("\n💀 GAME OVER 💀")
            print(f"🎯 The correct number was: {numrand}")

        guess += 1

    except ValueError:
        print("❌ Please enter a valid integer!\n")
```
