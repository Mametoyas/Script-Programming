## **Python คืออะไร?**

Python เป็นภาษาโปรแกรมระดับสูง (High-Level Programming Language) ที่ได้รับความนิยมอย่างมากในปัจจุบัน เนื่องจากมีรูปแบบการเขียนที่เรียบง่าย อ่านเข้าใจง่าย และสามารถนำไปประยุกต์ใช้ได้หลากหลายด้าน ตั้งแต่งานเขียนสคริปต์ขนาดเล็กไปจนถึงการพัฒนาระบบขนาดใหญ่

### **High-Level Language**

Python เป็นภาษาโปรแกรมระดับสูง (High-Level Language) ซึ่งหมายถึงภาษาที่มีรูปแบบการเขียนใกล้เคียงกับภาษามนุษย์ ทำให้ผู้พัฒนาสามารถอ่านและเขียนโปรแกรมได้ง่ายกว่าภาษาเครื่องหรือภาษาระดับต่ำ

### **Interpreted Language**

Python เป็นภาษาแบบ Interpreted Language หมายความว่าโปรแกรมจะถูกแปลและทำงานทีละคำสั่งผ่าน Interpreter โดยไม่จำเป็นต้อง Compile ทั้งโปรแกรมก่อนรัน ช่วยให้สามารถทดสอบและพัฒนาโปรแกรมได้อย่างรวดเร็ว

#### **Sample Snippets**

```python
# Snippet 1: Hello World!
# Concept: Basic print() function.

print("\033[1;36m")
print("╔════════════════════════════╗")
print("║       Hello, World!        ║")
print("╚════════════════════════════╝")
print("\033[0m")
```

```python
# Snippet 2: Variables and Basic Data Types
# Concept: int, float, str, bool

print("\n" + "=" * 70)
print("           PYTHON BASIC DATA TYPES DEMO")
print("=" * 70)

# =========================
# INTEGER
# =========================
age = 30

print("\nINTEGER")
print("-" * 30)
print("Value :", age)
print("Type  :", type(age).__name__)


# =========================
# FLOAT
# =========================
price = 19.99

print("\nFLOAT")
print("-" * 30)
print("Value :", price)
print("Type  :", type(price).__name__)


# =========================
# STRING
# =========================
name = "Alice"
greeting = "Hello"
message = "Python is fun!"

print("\nSTRING")
print("-" * 30)
print(f"{greeting}, {name}")
print("Message:", message)
print("Type   :", type(name).__name__)


# =========================
# MULTI-LINE STRING
# =========================
long_text = """This is a multi-line string.
It can span multiple lines.
Useful for documentation and notes."""

print("\nMULTI-LINE STRING")
print("-" * 30)
print(long_text)


# =========================
# BOOLEAN
# =========================
is_student = True
is_adult = False

print("\nBOOLEAN")
print("-" * 30)
print("Student:", is_student)
print("Adult  :", is_adult)
print("Type   :", type(is_student).__name__)


# =========================
# SUMMARY TABLE
# =========================
print("\n" + "=" * 70)
print("SUMMARY TABLE")
print("=" * 70)

print(f"{'Variable':<15} {'Value':<15} {'Type'}")
print("-" * 70)

variables = {
    "age": age,
    "price": price,
    "name": name,
    "is_student": is_student
}

for var_name, value in variables.items():
    print(f"{var_name:<15} {str(value):<15} {type(value).__name__}")

print("-" * 70)
print("Program completed successfully.")
print("=" * 70)
```

```python
# Snippet 3: User Input and Simple Arithmetic
# Concept: Using input() to get data from the user, type conversion, and basic arithmetic operations.

# Get user's name
user_name = input("Enter your name: ")
print("Hello,", user_name + "!")
print()

# Get age and convert to integer
# input() always returns a string, so we need to convert it to int for calculations
age_str = input("Enter your age: ")
user_age = int(age_str) # Convert string to integer
print("You are", user_age, "years old.")
print("In 5 years, you will be", user_age + 5, "years old.")
print()

# Get two numbers from the user for arithmetic
num1_str = input("Enter the first number: ")
num2_str = input("Enter the second number: ")

# Convert inputs to float to handle decimals
number1 = float(num1_str)
number2 = float(num2_str)

# Perform arithmetic operations
sum_result = number1 + number2
difference_result = number1 - number2
product_result = number1 * number2
quotient_result = number1 / number2 # Division results in a float

print("Sum:", sum_result)
print("Difference:", difference_result)
print("Product:", product_result)
print("Quotient:", quotient_result)
```

```python
# 1. การคำนวณพื้นที่สี่เหลี่ยม
print("--- การคำนวณพื้นที่สี่เหลี่ยม ---")
# รับค่าความกว้างจากผู้ใช้
width_str = input("กรุณาใส่ค่าความกว้างของสี่เหลี่ยม (หน่วย): ")
width = float(width_str)

# รับค่าความยาวจากผู้ใช้
length_str = input("กรุณาใส่ค่าความยาวของสี่เหลี่ยม (หน่วย): ")
length = float(length_str)

# คำนวณพื้นที่
area = width * length

# แสดงผลลัพธ์
print(f"สี่เหลี่ยมที่มีความกว้าง {width} หน่วย และความยาว {length} หน่วย")
print(f"มีพื้นที่เท่ากับ: {area:.2f} ตารางหน่วย")
```

```python
# 2. การแปลงอุณหภูมิเซลเซียสเป็นฟาเรนไฮต์
print("--- การแปลงอุณหภูมิเซลเซียสเป็นฟาเรนไฮต์ ---")
# รับค่าอุณหภูมิเซลเซียสจากผู้ใช้
celsius_str = input("กรุณาใส่อุณหภูมิเป็นองศาเซลเซียส (°C): ")
celsius = float(celsius_str)

# คำนวณอุณหภูมิฟาเรนไฮต์
# สูตร: F = (C * 9/5) + 32
fahrenheit = (celsius * 9/5) + 32

# แสดงผลลัพธ์
print(f"{celsius:.2f}°C เท่ากับ {fahrenheit:.2f}°F")
```
